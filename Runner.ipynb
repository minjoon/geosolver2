```python
from geosolver.utils.prep import paragraph_to_sentences
from geosolver.utils.prep import sentence_to_words_statements_values
sents = paragraph_to_sentences('In the figure shown above, AC = 6. What is the length of segment AB?')
sentence_words = dict()
sentence_expressions = dict()
for sent in sents:
    (sent_words, sent_exprs, x) = sentence_to_words_statements_values(sents[sent])
    sentence_words[sent] = sent_words
    sentence_expressions[sent] = sent_exprs
print sentence_words
print sentence_expressions


choices={1: '3', 2: '5', 3: '6', 4: '7', 5: 'It cannot be determined from the information given'}
choice_words = dict()
choice_expressions = dict()

for choice in choices:
    (choice_wds, choice_exprs, x) = sentence_to_words_statements_values(choices[choice])
    choice_words[choice] = choice_wds
    choice_expressions[choice] = choice_exprs

print choice_words
print choice_expressions

#{sent: sentence_to_words_statements_values(sents[sent]) for sent in sents}
```
